```python
import pandas as pd
import numpy as np
```

1)Read the csv file



```python
df=pd.read_csv("/content/data clean.csv.csv")
```

2)Check for Null values

```python
df.isnull().sum()
```

3) Remove null values from Age column replace it with mean value

```python
df['Age']
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Age']
```

4)Remove null values from Rating column,fill the null value with 0

```python
df['Rating'].fillna(0,inplace=True)
```

5) Data is now clean

```python
df.isnull().sum()
```

6) Age column data type and Salary column is format

```python
df['Age']=df['Age'].astype(int)
df['Salary']=df['Salary'].str.replace('$',' ',regex=True)
df['Salary']=df['Salary'].str.replace('k',' ',regex=True)
```

7)Location column is also formatted

```python

df.dtypes

```

```python
df['Location']=df['Location'].str.replace(',','')
df['Location']=df['Location'].str.replace('IndiaIn','India')
df['Location']=df['Location'].str.replace('New YorkNy','New York')
df['Location']=df['Location'].str.replace('Australia Aus','Australia')


df['Location']
```

8) Established column is formatted and data type is changed

```python
df['Established'] = df['Established'].astype(str)
df['Established']=df['Established'].str.replace('-1','0',)
```

```python

df['Established']
```

9)Easy Apply column have -1 we removed it by False

```python
df['Easy Apply']=df['Easy Apply'].astype(str)
df
```

10)Rating columns have negative values replaced with 0.0

```python
df['Rating']=df['Rating'].replace(-1.0,0.0)
df['Rating']
```
